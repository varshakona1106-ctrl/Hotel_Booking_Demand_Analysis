### Understanding "~" in Pandas

The "~" operator means "NOT" or "reverse the condition".

'df[condition]' keeps only the rows matching the condition, whereas 'df[~condition]' keeps all rows except those matching the condition.

"~condition = NOT these rows" else "Remove rows matching this condition."
